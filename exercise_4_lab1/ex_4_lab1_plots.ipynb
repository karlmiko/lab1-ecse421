### Working with accelerometer

```python
import numpy as np
from matplotlib import pyplot as plt
import scipy.integrate as it
```

```python
# read data file
with open('sample_acc.txt') as f:
    contents = f.readlines()
    f.close()
```

```python
# printing labels and first few lines
for line in contents[:5]:
    print(line)
```

```python
# create empty lists for X, Y, Z
X = []
Y = []
Z = []
```

```python
# splitting and getting data
for line in contents:
    list_line = line.split(" ")
    
    X_sample = float(list_line[3])
    Y_sample = float(list_line[5])
    Z_sample = float(list_line[7])
    
    X.append(X_sample)
    Y.append(Y_sample)
    Z.append(Z_sample)
```

```python
# making them in numpy to plot

Acc_X_np = np.array(X)
Acc_Y_np = np.array(Y)
Acc_Z_np = np.array(Z)
```

```python
# plot acceleration X, Y, Z

plt.figure(figsize=(20,16))

titles = ["X ", "Y ", "Z "]
positions = 0
x = np.array(list(range(0, len(X))))
sec = len(X)/10

for Acc in [Acc_X_np, Acc_Y_np, Acc_Z_np]:
    
    plt.subplot(3,1,positions+1)
    plt.plot(x, Acc)
    plt.title(titles[positions] + "Acceleration over " + str(sec) + " seconds"), plt.xlabel("Samples (10 Hz)"), plt.ylabel("g")
    positions += 1
        
plt.show()
```

```python
# find velocity from simple integration X, Y, Z

vel =0
d   =0
Vel_X = []

for i in range(len(X)-1):
    d=(X[i]+X[i+1])/2
    vel=vel+d
    Vel_X += [vel]
    
vel =0
d   =0
Vel_Y = []

for i in range(len(Y)-1):
    d=(Y[i]+Y[i+1])/2
    vel=vel+d
    Vel_Y += [vel]

vel =0
d   =0
Vel_Z = []

for i in range(len(Z)-1):
    d=(Z[i]+Z[i+1])/2
    vel=vel+d
    Vel_Z += [vel]
```

```python
# plot velocity X, Y, Z

plt.figure(figsize=(20,16))

titles = ["X ", "Y ", "Z "]
positions = 0
x = np.array(list(range(0, len(Vel_X))))
sec = len(Vel_X)/10

# Integral
Vel_X_np = np.array(Vel_X)
Vel_Y_np = np.array(Vel_Y)
Vel_Z_np = np.array(Vel_Z)

for Vel in [Vel_X_np, Vel_Y_np, Vel_Z_np]:
    
    plt.subplot(3,1,positions+1)
    plt.plot(x, Vel)
    plt.title(titles[positions] + "Velocity over " + str(sec) + " seconds"), plt.xlabel("Samples (10 Hz)"), plt.ylabel("g")
    positions += 1
        
plt.show()
```

```python
# find displacement from simple integration velocity X, Y, Z

dis =0
d   =0
Dis_X = []

for i in range(len(Vel_X)-1):
    d=(Vel_X[i]+Vel_X[i+1])/2
    dis=dis+d
    Dis_X += [dis]
    
dis =0
d   =0
Dis_Y = []

for i in range(len(Vel_Y)-1):
    d=(Vel_Y[i]+Vel_Y[i+1])/2
    dis=dis+d
    Dis_Y += [dis]

dis =0
d   =0
Dis_Z = []

for i in range(len(Vel_Z)-1):
    d=(Vel_Z[i]+Vel_Z[i+1])/2
    dis=dis+d
    Dis_Z += [dis]
```

# Moving Average

```python
# From above, X, Y, Z contain the unfiltered data
```

```python
def moving_average(L, N=5):
    new_L = []
    temp = []
    for i in range(len(L)-N):
        for j in range(N):
            temp+=[L[i+j]]
        A = sum(temp)/N
        new_L += [A]
        temp = []
    return new_L
```

```python
# N number of elements per calculation
N = 6
```

```python
len(X), len(Y), len(Z)
```

```python
Fil_X = np.array(moving_average(X, N=N))
Fil_Y = np.array(moving_average(Y, N=N))
Fil_Z = np.array(moving_average(Z, N=N))
```

```python
sec_f = Fil_X.shape[0]/10
sec = Acc_X_np.shape[0]/10
```

```python
# acceleration X, Y, Z, filtered by moving average compared to raw

plt.figure(figsize=(20,16))

titles = ["X ", "Y ", "Z "]
positions = 0
x = np.array(list(range(0, Fil_X.shape[0])))
x_a = np.array(list(range(0, Acc_X_np.shape[0])))

for Fil, Acc, title in zip([Fil_X, Fil_Y, Fil_Z], [Acc_X_np, Acc_Y_np, Acc_Z_np], titles):
    
    plt.subplot(3,2,positions+1)
    plt.plot(x, Fil)
    plt.title(title + "Acceleration over " + str(sec_f) + " seconds"), plt.xlabel("Samples (10 Hz)"), plt.ylabel("g")
    positions += 1
    plt.subplot(3,2,positions+1)
    plt.plot(x_a, Acc)
    plt.title(title + "Acceleration over " + str(sec) + " seconds"), plt.xlabel("Samples (10 Hz)"), plt.ylabel("g")
    positions += 1
        
plt.show()
```
